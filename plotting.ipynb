```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
x = np.linspace(0, 10, 1000)
# Refer From: 
# Python Data Science Handbook by Jake VanderPlas; Most of it content is available on GitHub.
```

```python
#Lines

fig = plt.figure(figsize=[13,7], frameon=False)
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':')  # dotted
```

```python
fig = plt.figure(figsize=[13,7], frameon=False)

plt.plot(x, np.sin(x - 0), color='blue')        
plt.plot(x, np.sin(x - 1), color='g')           
plt.plot(x, np.sin(x - 2), color='0.75')        
plt.plot(x, np.sin(x - 3), color='#FFDD44')     
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) 
plt.plot(x, np.sin(x - 5), color='chartreuse')
plt.plot(x, np.sin(x - 5), color='chartreuse')
```

```python
#Scatter plot

from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
print(iris.feature_names)
fig = plt.figure(figsize=[13,7], frameon=False)

plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```

```python
#Histogram 

df_tips = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
sns.set(font_scale=1.4)
df_tips['total_bill'].plot(kind='hist', figsize=(10, 10))
plt.xlabel("Total Bill Amount ($)", labelpad=14)
plt.ylabel("Frequency", labelpad=14)
plt.title("Distribution of Tip Bill Amounts ($)", y=1.015, fontsize=22)
```

```python
#Bar
fig=plt.gcf()
fig.set_size_inches(18.5,10.5)
sns.set(style='darkgrid')

data = pd.read_csv('matches.csv')
wins=pd.DataFrame(data['winner'].value_counts())
# print(wins.index)
wins['name']=wins.index
plt.xticks(rotation=90,fontsize=12)
plt.yticks(fontsize=16)
plt.bar(wins['name'],
        wins['winner'],
        color=['#15244C','#FFFF48','#292734','#EF2920','#CD202D','#ECC5F2',
               '#294A73','#D4480B','#242307','#FD511F','#158EA6','#E82865',
               '#005DB7','#C23E25','#E82865']
        ,alpha=0.8)
count=0
for i in wins['winner']:
    plt.text(count-0.15,i-4,str(i),size=15,color='black',rotation=90)
    count+=1
plt.title('Total wins by each team',fontsize=20)
plt.xlabel('Teams',fontsize=15)
plt.ylabel('Total no. of matches won(2008-2019)',fontsize=14)
plt.show()
```

```python
#Swarn

fig=plt.gcf()
fig.set_size_inches(18.5,10.5)
# print(data['win_by_runs'])
sns.swarmplot(data['season'],data[data['win_by_runs']!=0]['win_by_runs'],s=10)
plt.xticks(rotation=90,fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Season',fontsize=14)
plt.ylabel('Runs',fontsize=14)
plt.title('Season wise match summary of matches won by runs',fontsize=14)
plt.show()
```

```python
#Pie

matches = data
deliveries = pd.read_csv('deliveries.csv')
city_counts=matches.groupby('city').apply(lambda x:x['city'].count()).reset_index(name='Match Counts')
top_cities_order=city_counts.sort_values(by='Match Counts',ascending=False)
top_cities=top_cities_order[:20]
plt.figure(figsize=(9,9))
plt.pie(top_cities['Match Counts'],labels=top_cities['city'],autopct='%1.1f%%', startangle=30)
plt.axis('equal')
plt.title('Top Cities that have hosted IPL Matches',size=20)
```

```python
#Bar
fig=plt.gcf()
fig.set_size_inches(18.5,10.5)
sns.set(style='darkgrid')

data = pd.read_csv('top50.csv', encoding='iso-8859-1')
wins=pd.DataFrame(data['Artist.Name'].value_counts())
# print(wins.index)
wins['name']=wins.index
plt.xticks(rotation=90,fontsize=12)
plt.yticks(fontsize=16)
plt.bar(wins['name'],
        wins['Artist.Name'],
        color=['#15244C','#FFFF48','#292734','#EF2920','#CD202D','#ECC5F2',
               '#294A73','#D4480B','#242307','#FD511F','#158EA6','#E82865',
               '#005DB7','#C23E25','#E82865']
        ,alpha=0.8)
plt.title('Artist.Name vs occures',fontsize=20)
plt.xlabel('Artist.Name',fontsize=15)
plt.ylabel('Artist.Name Occures in Data Set',fontsize=14)
plt.show()
```
